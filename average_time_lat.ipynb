```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
ds = xr.open_dataset('ncdiffusion_hourly_pertransect.nc')   # data is plotted with temporal transect positions
```

```python
# for hannah time range 1
ds_clipped_time = ds.sel(time=slice('1984', '2004'))
lat_mask = (ds_clipped_time.lat >= 32.2538) & (ds_clipped_time.lat <= 35.6465)
ds_clipped = ds_clipped_time.where(lat_mask, drop=True)
overall_mean = ds_clipped['diffusivity'].mean().compute()
print(f"Mean Diffusivity (1984-2004, Region 1): {overall_mean.values:.4f}")
```
